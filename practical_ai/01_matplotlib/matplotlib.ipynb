```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
# adding title
plt.title("My Title")

# adding x and y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# adding limits
plt.xlim(0,8)
plt.ylim(0,16)

plt.plot(x,y)
plt.savefig('myfirstplot.jpg')
```
