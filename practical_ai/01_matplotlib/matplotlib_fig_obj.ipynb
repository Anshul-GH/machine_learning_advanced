```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
a = np.linspace(0,10,11)
```

```python
a
```

```python
b = a**4
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
```

```python
fig = plt.figure()

axes_large = fig.add_axes([0,0,1,1])
axes_large.plot(a,b)

axes_small = fig.add_axes([0.2,0.2,0.5,0.5])
axes_small.plot(x,y)

axes_tiny = fig.add_axes([0.4,0.4,0.1,0.1])
axes_tiny.plot(a,b)
```

```python
fig = plt.figure()

axes_large = fig.add_axes([0,0,1,1])
axes_large.plot(a,b)
axes_large.set_xlabel('X Large')
axes_large.set_ylabel('Y Large')
axes_large.set_xlim(0,10)
axes_large.set_ylim(0,10000)
axes_large.set_title('Large')

axes_small = fig.add_axes([0.2,0.2,0.5,0.5])
axes_small.plot(x,y)
axes_small.set_xlabel('X Small')
axes_small.set_ylabel('Y Small')
axes_small.set_xlim(0,9)
axes_small.set_ylim(0,18)
axes_small.set_title('Small')


axes_tiny = fig.add_axes([0.4,0.4,0.1,0.1])
axes_tiny.plot(a,b)
axes_tiny.set_xlabel('X Tiny')
axes_tiny.set_ylabel('Y Tiny')
axes_tiny.set_xlim(0,10)
axes_tiny.set_ylim(0,10000)
axes_tiny.set_title('Tiny')
```
